# Data preprocessing

```r
set.seed(111) # initialisation du générateur
# Extraction des échantillons
test.ratio <- .2   # part de l'échantillon test de 20%
npop <- nrow(rain_data) # nombre de lignes dans les données
nvar <- ncol(rain_data) # nombre de colonnes
# taille de l'échantillon test
ntest <- ceiling(npop * test.ratio) 
# indices de l'échantillon test
testi <- sample(1:npop, ntest)
#tirer aleatoire trainning sample
# indices de l'échantillon d'apprentissage
appri <- setdiff(1:npop, testi) 
```

```r
# construction de l'échantillon d'apprentissage
datappr <- rain_data[appri, -c(1,19)] 
#-11 veut dire enlever le colone 11
# construction de l'échantillon test
datestr <- rain_data[testi, -c(1,19)] 
# vérification
str(datappr)
print(" ")
str(datestr)
#summary(datappr)
```

```r
options(repr.plot.width = 8, repr.plot.height = 4)
# Définition d'une fonction pour un graphe coloré et des échelles fixes sur les
# axes
plot.res <- function(x, y, titre = "titre",Hmin=-20, Hmax=50, Vmin=-20, Vmax=40) {
    plot(x, y, col = "blue", xlim = c(Hmin, Hmax), ylim = c(Vmin, Vmax), ylab = "Résidus", 
        xlab = "Valeurs prédites", main = titre, pch = 20)
    # points(x2, y, col='red')
    abline(h = 0, col = "green")
}
```

```r
# NEW
V=10 ; nV=floor(nrow(datappr)/V)
S=sample(1:nrow(datappr),replace=FALSE)
error.CV = c()
for(v in 1:V)
{ # Rq : les deux dernières obs sont tjs dans l'échantillon d'apprentissage...
    datappr.learn=datappr[-c(S[(nV*(v-1)):(nV*v)]),] 
    datappr.valid=datappr[c(S[(nV*(v-1)):(nV*v)]),]
    error.CV=c(error.CV,mean((datappr.valid$rain-predict(aov(rain ~ ., data=datappr.learn),newdata=datappr.valid))^2))
}
mean(error.CV)

print(reg.lasso.cv)
```

## Modèle quadratique et sélection de variable

### AIC

```r
# Estimation du modèle de toute interaction d'ordre 2
reg.glm<- glm(rain ~ .^2, data = datappr)
# Recherche du meilleur modèle au sens 
# du critère d'Akaïke par méthode descendante
reg.glm.step <- step(reg.glm, direction = "backward")
```

```r
reg.glm.step
```

```r
# Coefficients du modèle
anova(reg.glm.step, test = "F")
#Il y a trop de paramètre
```

### LASSO

```r
# Comparer avec un modèle quadratique avec pénalité L1
x.mat2 <- model.matrix(rain ~ .^2 - 1, data = datappr)
reg.lasso2.cv <- cv.glmnet(y = datappr[, "rain"], x = x.mat2)
coef(reg.lasso2.cv, s = "lambda.1se")
```

### Comparaison AIC et LASSO

```r
# Extraction des valeurs ajustées et des résidus
fit.glm <- reg.glm.step$fitted.values
res.glm <- reg.glm.step$residuals
fit.lasso2 <- predict(reg.lasso2.cv, s = "lambda.min", newx = x.mat2)
res.lasso2 <- datappr$rain - fit.lasso2

# Graphe des résidus
options(repr.plot.width = 8, repr.plot.height = 8)
par(mfrow = c(2, 2))
plot.res(fit.lm, res.lm, "linéaire")
plot.res(fit.lasso, res.lasso, "linéaire, pénalité L1")
plot.res(fit.glm, res.glm, "quadratique, backward AIC")
plot.res(fit.lasso2, res.lasso2, "quadratique, pénalité L1")
```

## Prévision et l'erreur de régression

### Erreur de régression

#### Pour modèle quadratique

```r
# Erreur quadratique par modèle déterministe tp_arome
sum((datestr[,"tp_arome"] - datestr[,"rain"])^2) / nrow(datestr)
```

```r
TauxbonCl <- function(T) {
a=sum(T)
b=0
l=sqrt(length(T))
for(i in 1:l) {
    b=b+T[i,i]
}
monResultat <- b/a 
return(monResultat*100)
}
```

```r
# Calcul des prévisions pour le nomdèle quadratique backward AIC
pred.glm <- predict(reg.glm.step, newdata = datestr)
# Erreur quadratique moyenne de prévision (MSE)
sum((pred.glm - datestr[, "rain"])^2) / nrow(datestr)
```

On peut remarquer que notre modèle linéaire est mieux que celle de déterministe

```r
# Calcul des prévisions pour le modèle sans intéraction
pred.lm <- predict(reg.lm, newdata = datestr)
# Erreur quadratique moyenne de prévision (MSE)
sum((pred.lm - datestr[, "rain"])^2) / nrow(datestr)
```
