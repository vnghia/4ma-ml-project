# Data preprocessing

```r
set.seed(111) # initialisation du générateur
# Extraction des échantillons
test.ratio <- .2   # part de l'échantillon test de 20%
npop <- nrow(rain_data) # nombre de lignes dans les données
nvar <- ncol(rain_data) # nombre de colonnes
# taille de l'échantillon test
ntest <- ceiling(npop * test.ratio) 
# indices de l'échantillon test
testi <- sample(1:npop, ntest)
#tirer aleatoire trainning sample
# indices de l'échantillon d'apprentissage
appri <- setdiff(1:npop, testi) 
```

```r
# construction de l'échantillon d'apprentissage
datappr <- rain_data[appri, -c(1,19)] 
#-11 veut dire enlever le colone 11
# construction de l'échantillon test
datestr <- rain_data[testi, -c(1,19)] 
# vérification
str(datappr)
print(" ")
str(datestr)
#summary(datappr)
```

```r
# construction de l'échantillon d'apprentissage
datappq <- rain_data[appri, -c(1,18)] 
#-11 veut dire enlever le colone 11
# construction de l'échantillon test
datestq <- rain_data[testi, -c(1,18)] 
# vérification
str(datappq)
print(" ")
str(datestq)
#summary(datappr)
```

```r
options(repr.plot.width = 8, repr.plot.height = 4)
# Définition d'une fonction pour un graphe coloré et des échelles fixes sur les
# axes
plot.res <- function(x, y, titre = "titre",Hmin=-20, Hmax=50, Vmin=-20, Vmax=40) {
    plot(x, y, col = "blue", xlim = c(Hmin, Hmax), ylim = c(Vmin, Vmax), ylab = "Résidus", 
        xlab = "Valeurs prédites", main = titre, pch = 20)
    # points(x2, y, col='red')
    abline(h = 0, col = "green")
}
```

# Linear Model

## Modèle sans interaction et avec intéraction

```r
# estimation du modèle sans interaction
reg.lm <- aov(rain ~1+td+ff, data = datappr)
# ~. c'est sans intéraction,^2 est avec intéraction

# Extraction des résidus et des valeurs ajustées de ce modèle
res.lm <- reg.lm$residuals
fit.lm <- reg.lm$fitted.values

# Graphe des résidus. 
plot.res(fit.lm,res.lm,"Régression linéaire sans sélection de variables",-10,20,-15,35)
```

```r
summary(reg.lm)
plot(reg.lm)
```

```r
# Graphe des résidus du modèle AROME
par(mfrow = c(1, 2))
plot.res(datappr[, "tp_arome"],
         datappr[, "tp_arome"] - datappr[, "rain"], "linéaire, rain seul",-10,100,-50,100)
plot.res(fit.lm, res.lm, "Linéaire, sans sélection",-10,20,-15,30)
par(mfrow = c(1, 1))
```

## Lasso selection

```r
library(glmnet)
# avec des variables quantitatives seulement
reg.lasso.quanti <- glmnet(y = datappr[, 17],
                           x = as.matrix(datappr[, -c(1, 17)]))
# avec toutes les variables, créer d'abord la matrice d'expériences 
# avec 'model.matrix' (penser à retirer l'intercept du modèle)
x.mat <- model.matrix(rain ~ . - 1, data = datappr)
reg.lasso <- glmnet(y = datappr$rain, x = x.mat)
options(repr.plot.width = 15, repr.plot.height = 20)
plot(reg.lasso, xvar = "lambda", label = TRUE)
#legend("topright", 
#       legend = paste(1:ncol(x.mat), " - ", colnames(x.mat)))
```

```r
summary(reg.lasso)
```

```r
head(x.mat)
```

```r
# choix du paramètre de régularisation par validation croisée
reg.lasso.cv <- cv.glmnet(y = datappr[, 17], x = x.mat)
plot(reg.lasso.cv)
# c'est bizarre parce que lambda optimal => modèle de 1 variable, connecté avec la transformation de distribution?
```

```r
# valeur estimée
paste("CV estimate of lambda :", round(reg.lasso.cv$lambda.1se, 3))
# modèle correspondant
coef(reg.lasso.cv, s = "lambda.1se")
```

```r
# NEW : 
plot(reg.lasso, xvar = "lambda", label = TRUE,xlim=c(-1,1),ylim=c(-1,1))
abline(v=log(reg.lasso.cv$lambda.1se),col="red")
```

```r
# NEW : 
# valeur estimée
paste("CV estimate of lambda :", round(reg.lasso.cv$lambda.min, 3))
# modèle correspondant
coef(reg.lasso.cv, s = "lambda.min")

plot(reg.lasso, xvar = "lambda", label = TRUE,xlim=c(-2,0),ylim=c(-5,40))
abline(v=log(reg.lasso.cv$lambda.min),col="red")
```

```r
# Extraction des valeurs ajustées et des résidus
fit.lasso <- predict(reg.lasso.cv, s = "lambda.min", newx = x.mat)
res.lasso <- datappr$rain - fit.lasso
# Graphe des résidus
options(repr.plot.width = 8, repr.plot.height = 4)
par(mfrow = c(1, 2))
plot.res(fit.lm, res.lm, "linéaire")
plot.res(fit.lasso, res.lasso, "linéaire, pénalité L1")
```

```r
# Extraction des valeurs ajustées et des résidus

fit.lasso <- predict(reg.lasso.cv, s = "lambda.min", newx = x.mat)
res.lasso <- datappr$rain - fit.lasso

fit.lasso.1se <- predict(reg.lasso.cv, s = "lambda.1se", newx = x.mat) # NEW
res.lasso.1se <- datappr$rain - fit.lasso.1se # NEW

# Graphe des résidus
options(repr.plot.width = 12, repr.plot.height = 4)
par(mfrow = c(1, 3))
plot.res(fit.lm, res.lm, "Linéaire, sans sélection",-10,20,-15,35)
plot.res(fit.lasso, res.lasso, "Linéaire, pénalité L1, lambda min",-10,20,-15,35)
plot.res(fit.lasso.1se, res.lasso.1se, "Linéaire, pénalité L1, lambda 1se",-10,20,-15,35) # NEW
```

```r
# NEW : 
paste("Modèle linéaire sans séletion:",mean(res.lm^2))
paste("LASSO avec lambda.min:",mean(res.lasso^2))
paste("LASSO avec lambda.1se:",mean(res.lasso.1se^2))
```

```r
# NEW
V=10 ; nV=floor(nrow(datappr)/V)
S=sample(1:nrow(datappr),replace=FALSE)
error.CV = c()
for(v in 1:V)
{ # Rq : les deux dernières obs sont tjs dans l'échantillon d'apprentissage...
    datappr.learn=datappr[-c(S[(nV*(v-1)):(nV*v)]),] 
    datappr.valid=datappr[c(S[(nV*(v-1)):(nV*v)]),]
    error.CV=c(error.CV,mean((datappr.valid$rain-predict(aov(rain ~ ., data=datappr.learn),newdata=datappr.valid))^2))
}
mean(error.CV)

print(reg.lasso.cv)
```
